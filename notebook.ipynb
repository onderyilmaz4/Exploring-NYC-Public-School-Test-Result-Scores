Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a maximum score of 800 points. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()


## Question 1 - Best Math Schools

# Creating a new column that indicates the percentage of success of the schools
schools["score_percentage"] = schools["average_math"]* 100/800

# Finding the schools that scored above 80% percent in SAT
school_subset_80 = schools[schools["score_percentage"] >= 80]

# Finding the best schools in math sorted in descending order
best_math_schools = school_subset_80[["school_name", "average_math"]].sort_values(by = "average_math", ascending = False)

print(best_math_schools)

## Question 2 - Top 10 Performing Schools

# Creating a new column that displays the total scores across the three SAT sections
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

# Sorting the schools according to "total_SAT" and subsetting using "school_name" and "total_SAT" columns
school_subset_total_score = schools[["school_name", "total_SAT"]].sort_values(by = "total_SAT", ascending = False)

# Finding the most successful 10 schools
top_10_schools = school_subset_total_score.iloc[0:10]
print(top_10_schools)

## Question 3 - NYC Borough with the Largest SD

# Grouping the schools by "borough"s they belong, calculating the number of the schools in the "borough"s and the mean and the standard deviation of their "total_SAT" values, rounding the results to 2 decimals
borough_table = schools.groupby("borough")["total_SAT"].agg(["count","mean", "std"]).round(2).sort_values(by = "std", ascending = False)

# Finding the borough with largest sd
condition_std = (borough_table["std"] == borough_table["std"].max())
largest_std_dev =  borough_table[condition_std]

# Renaming the columns
largest_std_dev = largest_std_dev.rename(columns={"count":"num_schools", "mean":"average_SAT", "std":"std_SAT"})
```
